```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('insurance_pre.csv')
```

```python
dataset
```

```python
dataset
dataset=pd.get_dummies(dataset,drop_first=True)
```

```python
dataset
```

```python
indep = dataset[['age','bmi', 'children', 'sex_male','smoker_yes']]
dep=dataset['charges']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(indep, dep, test_size=0.30, random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVR
param_grid={'kernel':["rbf",'poly','sigmoid','linear'], 'C':[10,100,1000,2000,3000], 'gamma':['auto','scale']}
grid=GridSearchCV(SVR(), param_grid, refit=True, verbose=3, n_jobs=-1)
grid.fit(X_train, y_train)
```

```python
GridSearchCV(estimator=SVR(), n_jobs=-1, param_grid={'C':[10,100,1000,2000,3000],'gamma':['auto','scale'],'kernel':['rbf','poly','sigmoid','linear']},verbose=3)
```

```python
re=grid.cv_results_
print("The R_score value for best parameter {}:".format(grid.best_params_))
```

```python
table=pd.DataFrame.from_dict(re)
```
